```python
import pandas as pd

df = pd.read_csv("simulated-data.csv")


```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier

vectorizer = CountVectorizer()
xgb = XGBClassifier()

train = df.sample(n = 200)

pipeline = Pipeline([('vectorizer', vectorizer), ('xgb', xgb)])

X = train['text']
y = train['is_electronics']

pipeline.fit(X, y)
df["training_score"] = pipeline.predict_proba(df['text'])[:, 1]

df.to_csv("simulated-data.csv")
```
